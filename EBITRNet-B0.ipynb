```python
# ==============================================================
# EfficientNet‑B0 + BiT‑ResNetv2 (channel‑wise attention with 0.8 dropout)
# ‑ train/test split (80%/20%)
# ‑ per‑epoch training metrics: loss & acc
# ‑ final curves (train + val) + classification report
# ==============================================================

import os, random, copy, time
import torch, timm
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, random_split
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

# ---------------- USER CONFIG ----------------
DATA_DIR   = '/kaggle/input/papaya-sugarcane-9026/Papaya-sugarcane dataset'
IMG_SIZE   = 224
BATCH_SIZE = 32
EPOCHS     = 50
LR         = 3e-4
TEST_FRAC  = 0.20
SEED       = 123
DEVICE     = 'cuda' if torch.cuda.is_available() else 'cpu'
# ---------------------------------------------

random.seed(SEED); torch.manual_seed(SEED)

# 1. Data ------------------------------------------------------------------
train_tfms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225]),
])
eval_tfms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225]),
])

full_ds = ImageFolder(DATA_DIR, transform=train_tfms)
n_total = len(full_ds)
n_test  = int(n_total * TEST_FRAC)
n_train = n_total - n_test

train_ds, test_ds = random_split(full_ds, [n_train, n_test], generator=torch.Generator().manual_seed(SEED))
test_ds.dataset.transform = eval_tfms

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  num_workers=4, pin_memory=True)
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, num_workers=4, pin_memory=True)

class_names = full_ds.classes
num_classes = len(class_names)
print(f'Train/Test: {n_train}/{n_test}  |  Classes: {num_classes}')

# 2. Model -----------------------------------------------------------------
class EfficientNet_BiT_Attention_Dropout(nn.Module):
    def __init__(self, n_classes):
        super().__init__()
        self.effnet = timm.create_model('efficientnet_b0', pretrained=True, num_classes=0, global_pool='')
        self.dropout = nn.Dropout(0.8)
        self.eff_gap = nn.AdaptiveAvgPool2d(1)

        self.bit = timm.create_model('resnetv2_50x1_bit.goog_in21k_ft_in1k', pretrained=True, num_classes=0, global_pool='')
        self.dropout = nn.Dropout(0.8)
        self.bit_gap = nn.AdaptiveAvgPool2d(1)

        self.attn_fc = nn.Sequential(
            nn.Linear(self.bit.num_features, 1280),
            nn.Sigmoid()
        )

        self.dropout = nn.Dropout(0.8) 
        self.dropout = nn.Dropout(0.8)   # Corrected to 0.8 as per title
        self.fc = nn.Linear(1280, n_classes)

    def forward(self, x):
        e_feat = self.eff_gap(self.effnet(x)).flatten(1)
        b_feat = self.bit_gap(self.bit(x)).flatten(1)
        attn_mask = self.attn_fc(b_feat)
        attn_feat = e_feat * attn_mask
        attn_feat = self.dropout(attn_feat)
        return self.fc(attn_feat)

model = EfficientNet_BiT_Attention_Dropout(num_classes).to(DEVICE)

# 3. Optimiser & loss ------------------------------------------------------
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=LR, weight_decay=1e-4)
scaler    = torch.cuda.amp.GradScaler(enabled=(DEVICE=='cuda'))

# 4. Train / eval helpers --------------------------------------------------
def step(loader, train=True):
    model.train(train)
    losses, preds, tgts = [], [], []
    for imgs, labels in loader:
        imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
        with torch.set_grad_enabled(train), \
             torch.cuda.amp.autocast(enabled=(DEVICE=='cuda')):
            outputs = model(imgs)
            loss    = criterion(outputs, labels)
        if train:
            optimizer.zero_grad(set_to_none=True)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        losses.append(loss.item())
        preds.extend(outputs.argmax(1).cpu().numpy())
        tgts .extend(labels.cpu().numpy())
    return sum(losses)/len(losses), accuracy_score(tgts, preds)

# 5. Training loop ---------------------------------------------------------
history = {k: [] for k in ['train_acc','train_loss','val_acc','val_loss']}
best_test, best_wts = 0, None
start_time = time.time()

for epoch in range(1, EPOCHS+1):
    tr_loss, tr_acc = step(train_loader, train=True)
    ts_loss, ts_acc = step(test_loader,  train=False)

    history['train_acc'].append(tr_acc)
    history['train_loss'].append(tr_loss)
    history['val_acc'].append(ts_acc)
    history['val_loss'].append(ts_loss)

    if ts_acc > best_test:
        best_test, best_wts = ts_acc, copy.deepcopy(model.state_dict())

    print(f'\nEpoch {epoch:03}/{EPOCHS}'
          f' | Train: {tr_acc*100:6.2f}% / {tr_loss:.4f}'
          f' | Test: {ts_acc*100:6.2f}% / {ts_loss:.4f}')

elapsed = (time.time() - start_time) / 60
print(f'\nTraining finished in {elapsed:.1f} min')

# Load best weights
model.load_state_dict(best_wts or model.state_dict())

# 6. Final classification report ------------------------------------------
all_preds, all_labels = [], []
model.eval()
with torch.no_grad():
    for imgs, labels in test_loader:
        preds = model(imgs.to(DEVICE)).argmax(1).cpu()
        all_preds.extend(preds.numpy())
        all_labels.extend(labels.numpy())

print('\nClassification report (TEST):')
print(classification_report(all_labels, all_preds, target_names=class_names, digits=2))

# 7. Training + Validation Curves (no markers) -----------------------------
epochs = range(1, EPOCHS+1)
plt.figure(figsize=(12, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(epochs, history['train_acc'], label='Train Accuracy')
plt.plot(epochs, history['val_acc'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train vs Validation Accuracy')
plt.legend()
plt.grid(False)

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(epochs, history['train_loss'], label='Train Loss')
plt.plot(epochs, history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train vs Validation Loss')
plt.legend()
plt.grid(False)

plt.tight_layout()
plt.show()

# 8. Save best model -------------------------------------------------------
torch.save(best_wts, 'efficientnetb0_bit_attention_best.pth')
print('\nBest model weights saved to efficientnetb0_bit_attention_best.pth')
```
